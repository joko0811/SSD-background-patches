```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,4,8,16,32,64,128])
tp_y = np.array([2946,2924,2879,2764,2587,2203,241,0])/3000
fp_y = np.array([39304,78038,156449,313013,628486,1261824,2242605,2249945])/3000
```

```python
fig =plt.figure()

fig.supxlabel("total number of patches")
fig.supylabel("total number of detection")

ax1 =fig.add_subplot(1,2,1)
ax1.set_title("TP detection")
ax1.plot(x,tp_y)
ax1.plot(x,tp_y,"ro")

ax2= fig.add_subplot(1,2,2)
ax2.set_title("FP detection")
ax2.plot(x,fp_y)
ax2.plot(x,fp_y,"ro")

fig.show()
```
