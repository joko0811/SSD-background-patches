```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_similarity_distribution(y_true,y_score):
    y_tp=y_score[y_true>0]
    y_fp=y_score[y_true<1]
    y_tp_xy = np.unique(np.round(y_tp,decimals=1), return_counts=True)
    y_fp_x, y_fp_y = np.unique(np.round(y_fp,decimals=1), return_counts=True)
    y_fp_x = np.append(y_fp_x, 1.0)
    y_fp_y = np.append(y_fp_y, 0)
    plt.plot(y_tp_xy[0], y_tp_xy[1], color='r')
    plt.plot(y_fp_x, y_fp_y, color='b')
    plt.xlim(0., 1.0)
    plt.show()
    plt.clf()
```

```python
g_npy_path="/home/okano/dev/SSD-background-patches/outputs/gb_result/"
gy_true=np.load(os.path.join(g_npy_path,"y_true.npy"))
gy_score=np.load(os.path.join(g_npy_path,"y_score.npy"))
```

```python
plot_similarity_distribution(gy_true, gy_score)
```

```python
n_npy_path="/home/okano/dev/SSD-background-patches/outputs/noise_result/"
ny_true=np.load(os.path.join(n_npy_path,"y_true.npy"))
ny_score=np.load(os.path.join(n_npy_path,"y_score.npy"))
```

```python
a_npy_path="/home/okano/dev/SSD-background-patches/outputs/adv_result/"
ay_true=np.load(os.path.join(a_npy_path,"y_true.npy"))
ay_score=np.load(os.path.join(a_npy_path,"y_score.npy"))
```
