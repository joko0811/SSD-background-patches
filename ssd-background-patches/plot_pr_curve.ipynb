```python
import os
import numpy as np
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay
from sklearn.metrics import average_precision_score
```

```python
os.getcwd()
```

```python
def plot_prcurve(y_true,y_score):
    precision,recall,threshold = precision_recall_curve(y_true,y_score)
    disp = PrecisionRecallDisplay(precision,recall)
    disp.plot()
```

```python
a_npy_path="/home/okano/dev/SSD-background-patches/outputs/adv_result/"
ay_true=np.load(os.path.join(a_npy_path,"y_true.npy"))
ay_score=np.load(os.path.join(a_npy_path,"y_score.npy"))
a_ap=average_precision_score(ay_true,ay_score)
a_ap
```

```python
plot_prcurve(ay_true,ay_score)
```
